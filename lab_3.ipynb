```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
x = np.array([1, 2, 3, 4])  # weeks
y = np.array([1, 3, 4, 8])  # corresponding y values

# Reshape x to a column vector and add a column of ones for the intercept term (bias)
X = np.vstack([np.ones(len(x)), x]).T

# Calculate the coefficients using the normal equation: (X^T * X)^-1 * X^T * y
theta = np.linalg.inv(X.T @ X) @ X.T @ y

# Print the result
print(f"Intercept (theta_0): {theta[0]}")
print(f"Slope (theta_1): {theta[1]}")

# Predict for week 5
week_to_predict = 5
y_pred = theta[0] + theta[1] * week_to_predict
print(f"Predicted value for week {week_to_predict}: {y_pred}")

# Plot the data points
plt.scatter(x, y, color='red', label='Actual data')

# Plot the regression line
x_line = np.linspace(min(x), week_to_predict, 100)
y_line = theta[0] + theta[1] * x_line
plt.plot(x_line, y_line, color='blue', label='Fitted Line')

# Highlight the predicted point
plt.scatter(week_to_predict, y_pred, color='green', marker='o', label=f'Prediction (Week {week_to_predict})')

# Labels and title
plt.xlabel("Weeks")
plt.ylabel("Values")
plt.title("Linear Regression Fit")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
x = np.array([1, 2, 3, 4])  # weeks
y = np.array([1, 3, 4, 8])  # corresponding y values

# Compute mean values
x_mean = np.mean(x)
y_mean = np.mean(y)

# Compute slope (theta_1)
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
theta_1 = numerator / denominator

# Compute intercept (theta_0)
theta_0 = y_mean - theta_1 * x_mean

# Print the result
print(f"Intercept (theta_0): {theta_0}")
print(f"Slope (theta_1): {theta_1}")

# Predict for week 5
week_to_predict = 8
y_pred = theta_0 + theta_1 * week_to_predict
print(f"Predicted value for week {week_to_predict}: {y_pred}")

# Compute Mean Squared Error (MSE)
y_predicted = theta_0 + theta_1 * x
mse = np.mean((y - y_predicted) ** 2)
print(f"Mean Squared Error: {mse}")

# Plot the data points
plt.scatter(x, y, color='red', label='Actual data')

# Plot the regression line
x_line = np.linspace(min(x), week_to_predict, 100)
y_line = theta_0 + theta_1 * x_line
plt.plot(x_line, y_line, color='blue', label='Fitted Line')

# Highlight the predicted point
plt.scatter(week_to_predict, y_pred, color='green', marker='o', label=f'Prediction (Week {week_to_predict})')

# Labels and title
plt.xlabel("Weeks")
plt.ylabel("Values")
plt.title("Linear Regression Fit")
plt.legend()
plt.grid()
plt.show()
```
